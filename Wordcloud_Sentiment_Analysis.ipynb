```
#I want to findout the most popular baby name from 1880 - 2009
#Now we do some data transformation and clean the dataframe for our answer
#I will create a new data table just the columns for my wordcloud called "Babynames2"
Babynames2=Babynames ['name']
```

```
#Import wordcloud libraries
#STOPWORDS is a function that automatically are removed from our results
#Most commonly imported as a seperate file and joined with the data frame
from wordcloud import WordCloud, STOPWORDS
wordcloud=WordCloud(
                        background_color='white',
                        stopwords=STOPWORDS,
                        max_words=200,
                        max_font_size=100,
                        random_state=42
                        ).generate(str(Babynames2))

#Display the generated image:
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

```
#I notice the Column name of my data frome is showing in my wordcloud
#So now I want to remove it
#Use the code below to rename columns and this case I want to remove it to nothing so I use
Babynames3=Babynames2.rename(columns={'name':''}, inplace = True)
```

```
#Import wordcloud libraries
#STOPWORDS is a function that automatically are removed from our results
#Most commonly imported as a seperate file and joined with the data frame
from wordcloud import WordCloud, STOPWORDS
wordcloud=WordCloud(
                        background_color='white',
                        stopwords=STOPWORDS,
                        max_words=200,
                        max_font_size=80,
                        random_state=0
                       ).generate(str(Babynames3))

#Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

```
#Lets take it one step further and say I want to Skew the data a little bit and replace William with Tim
Babynames4=Babynames3.replace('William', 'Tim')
```

```
#Import wordcloud libraries
#STOPWORDS is a function that removes words from our results
#Most commonly imported as a seperate file and joined with the data frame
from wordcloud import WordCloud, STOPWORDS
wordcloud=WordCloud(
                        background_color='white',
                        stopwords=STOPWORDS,
                        max_words=200,
                        max_font_size=80,
                        random_state=0
                       ).generate(str(Babynames4))

#Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

```
#I now want to remove object, length, and dtype from the word cloud by using the STOPWORDS function
from wordcloud import WordCloud, STOPWORDS
wordcloud=WordCloud(
                        background_color='white',
                        stopwords=['object', 'length', 'dtype'],
                        max_words=200,
                        max_font_size=80,
                        random_state=0
                       ).generate(str(Babynames4))

#Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
