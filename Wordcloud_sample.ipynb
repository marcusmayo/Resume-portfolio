```
#I start with loading all necessary libraries
import numpy as np
#Numpy provides fast mathematical function processing
import pandas as pd
#Pandas is used for easy indexing and data collection tools
from PIL import Image
#Pillow is a very good Python Image Library
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
#I am importing necessary packages for word cloud
import matplotlib.pyplot as plt #Matplotlib is used for data visualizations
% matplotlib inline
```

```
#Second, I manually add words from my resume
text=("Marcus Mayo Marcus Mayo Marcus Mayo Marcus Mayo Marcus Mayo Marcus Mayo Python Python Docker Docker Docker Docker Docker Docker Devops Devops Devops Devops Devops Devops Devops Linux Linux AWS AWS AWS AWS AWS AWS AWS AWS Cloud Computing Cloud Computing Cloud Computing Cloud Computing Cloud Computing Cloud Computing Git Git Git Git Ansible Ansible Ansible Ansible Jenkins Jenkins Jenkins Jenkins")

#Third I create the wordcloud project 
wordcloud= WordCloud(width=512, height=480, margin=0).generate(text)

#Fourth I insert the cloud to display the genereated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=30, y=0)
plt.show()
```
